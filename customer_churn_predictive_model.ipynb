```python
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline
import sklearn
```

```python
crm = pd.read_excel('dflp_case_sample_data.xlsx' , sheet_name='dflp_case_sample_data')
crm['label'] = crm['label'].astype(np.int64)
```

```python
crm_features = crm[['days_last_order' , 'total_orders' , 'unique_days_visited' , 'total_sessions' , 'label']]
crm_features
```

```python
#X = crm_features[['total_orders', 'total_sessions']]
#X = crm_features.drop('label', axis=1)
#y = crm_features.label
y = crm['label']
X = crm[['unique_days_visited' , 'total_sessions']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_valid,y_train,y_valid = train_test_split(X, y, random_state = 101, stratify=y, test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_model = DecisionTreeClassifier(random_state=10)
```

```python
dt_model.fit(X_train, y_train)
```

```python
dt_model.score(X_train, y_train)
```

```python
dt_model.score(X_valid, y_valid)
```
